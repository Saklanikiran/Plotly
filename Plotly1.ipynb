```python

import seaborn as sns
import plotly.express as px

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Plot the scatter plot using Plotly express
fig = px.scatter(titanic, x='age', y='fare', title='Scatter Plot of Age vs Fare in Titanic Dataset')
fig.show()
```

# Ans : 2

```python
# Load the tips dataset from Plotly
tips = px.data.tips()

# Plot the box plot using Plotly express
fig = px.box(tips, y='total_bill', title='Box Plot of Total Bill in Tips Dataset')
fig.show()
```

# Ans : 3

```python
# Plot the histogram using Plotly express
fig = px.histogram(tips, x='sex', y='total_bill', color='day', pattern_shape='smoker', title='Histogram of Total Bill by Sex with Smoker Pattern and Day Color')
fig.show()
```

# Ans : 4

```python
# Load the iris dataset from Plotly
iris = px.data.iris()

# Plot the scatter matrix plot using Plotly express
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color='species', title='Scatter Matrix of Iris Dataset')
fig.show()
```

# Ans : 5

```python
# A distplot is a combination of a histogram and a kernel density estimate (KDE) plot. It is used to visualize the distribution of a dataset.

# Plot a distplot using Plotly express
fig = px.histogram(tips, x='total_bill', marginal="rug", title='Distribution Plot of Total Bill')
fig.show()
```
